```python
import os
import sys
import glob
import numpy as np
import pickle
```

```python
import time
tstamp = int(time.time())

src_dir = "/mnt/data/yanzi/xiaomi_vacuum_as_data_collector/data_parsed_all"
dst_dir = "/home/gomezp/rss-data/xiaomi_all_{}".format(tstamp)

if not os.path.exists(dst_dir):
    os.makedirs(dst_dir)
```

```python
def normalize_loc(l):
    return (l - 32.) / 64.
```

```python
in_files = glob.glob("{}/*.pickle".format(src_dir))
for idx, in_fname in enumerate(in_files):
    parts = in_fname.split('/')[-1].split('_')
    date, time, dev, mac, _, pkt_type, _ = parts[:7]
    
    with open(in_fname, 'rb') as f:
        data = pickle.load(f)
    
    if data[1] is None:
        continue
    rss, x, y = data[0], *normalize_loc(np.array(data[1]))
    
    out_fname = "{}/img_{:.2f}_{:.2f}_{}_{}.magic{}.pickle".format(dst_dir, x, y, 2.0, -10, idx)
    
    with open(out_fname, 'wb') as f:
        print(out_fname)
        pickle.dump(rss, f)
```
